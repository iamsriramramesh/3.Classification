```python
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
dataset["classification_yes"].value_counts()
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dependent=dataset[['classification_yes']]


from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size=1/3, random_state=0)


from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
#LogisticRegression

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

param_grid = {
    'penalty': ['l1', 'l2'],
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'lbfgs', 'sag'],
    'class_weight': [None, 'balanced'],
    'max_iter': [100, 200, 300]
}

grid_LogisticRegression = GridSearchCV(LogisticRegression(), param_grid, refit=True, verbose=3, n_jobs=-1, scoring='f1_weighted')
grid_LogisticRegression.fit(x_train, y_train)

re_LogisticRegression=grid_LogisticRegression.cv_results_
grid_pred = grid_LogisticRegression.predict(x_test) 

from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test,grid_pred)
print("The confusion Matrix:\n",cm)

# print classification report 
from sklearn.metrics import classification_report
clf_report= classification_report(y_test,grid_pred)
print("The report:\n",clf_report)

from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,grid_pred,average='weighted')
print("The f1_macro value for best parameter{} :".format(grid_LogisticRegression.best_params_),f1_macro)

from sklearn.metrics import roc_auc_score
roc_auc_score=roc_auc_score(y_test,grid_LogisticRegression.predict_proba(x_test)[:,1])
print ("roc_auc_score=",roc_auc_score)
```

```python
#Naive Bayes
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import GridSearchCV

param_grid = {
    'var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6, 1e-5]
}

grid_GaussianNB = GridSearchCV(GaussianNB(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted')
# fitting the model for grid search 
grid_GaussianNB.fit(x_train, y_train) 

# print best parameter after tuning 
#print(grid.best_params_) 
re_GaussianNB=grid_GaussianNB.cv_results_
#print(re)
grid_predictions = grid_GaussianNB.predict(x_test) 
   
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, grid_predictions)
print("The confusion Matrix:\n",cm)

# print classification report 
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, grid_predictions)
print("The report:\n",clf_report)

from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,grid_predictions,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid_GaussianNB.best_params_),f1_macro)

from sklearn.metrics import roc_auc_score
roc_auc_score=roc_auc_score(y_test,grid_GaussianNB.predict_proba(x_test)[:,1])
print ("roc_auc_score=",roc_auc_score)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree'],
    'leaf_size': [10, 20, 30, 40],
    'p': [1, 2]  # For Minkowski metric
}

grid_KNeighborsClassifier = GridSearchCV(
    KNeighborsClassifier(), param_grid, refit=True, verbose=3, n_jobs=-1, scoring='f1_weighted'
)

# Fit the model (corrected indentation)
grid_KNeighborsClassifier.fit(x_train, y_train)

# Print best parameters (uncommented)
print("Best parameters:", grid_KNeighborsClassifier.best_params_)

# Make predictions
grid_predictions = grid_KNeighborsClassifier.predict(x_test)

# Evaluate performance
from sklearn.metrics import confusion_matrix, classification_report, f1_score, roc_auc_score

cm = confusion_matrix(y_test, grid_predictions)
clf_report = classification_report(y_test, grid_predictions)
f1_macro = f1_score(y_test, grid_predictions, average='weighted')

print("The confusion Matrix:\n", cm)
print("The report:\n", clf_report)
print("The f1_macro value for best parameter {}:".format(grid_KNeighborsClassifier.best_params_), f1_macro)

roc_auc_score = roc_auc_score(y_test, grid_KNeighborsClassifier.predict_proba(x_test)[:, 1])
print("roc_auc_score=", roc_auc_score)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

params = {
    'kernel': ['linear', 'rbf', 'poly', 'sigmoid'],
    'gamma': ['auto', 'scale'],
    'C': [0.1, 1, 10, 100],  # Added C parameter
    'class_weight': [None, 'balanced'],
    'decision_function_shape': ['ovo', 'ovr'],
    'degree': [2, 3, 4], 
}

grid_SVC = GridSearchCV(SVC(probability=True), params, refit=True, verbose=3, n_jobs=-1, scoring='f1_weighted')

# Fit the model (corrected indentation)
grid_SVC.fit(x_train, y_train)

# Print best parameters (uncommented)
print("Best parameters:", grid_SVC.best_params_)

# Make predictions
grid_predictions = grid_SVC.predict(x_test)

# Evaluate performance
from sklearn.metrics import confusion_matrix, classification_report, f1_score, roc_auc_score

cm = confusion_matrix(y_test, grid_predictions)
clf_report = classification_report(y_test, grid_predictions)
f1_macro = f1_score(y_test, grid_predictions, average='weighted')

print("The confusion Matrix:\n", cm)
print("The report:\n", clf_report)
print("The f1_macro value for best parameter {}:".format(grid_SVC.best_params_), f1_macro)

roc_auc_score = roc_auc_score(y_test, grid_SVC.predict_proba(x_test)[:, 1])
print("roc_auc_score=", roc_auc_score)
```

```python
#DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV


param_grid = {
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random'],
    'max_depth': [3, 5, 7, 9, 11, 13, 15], 
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2'],
    'class_weight': [None, 'balanced']
}

grid_DecisionTreeClassifier = GridSearchCV(DecisionTreeClassifier(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted')

# fitting the model for grid search 
grid_DecisionTreeClassifier.fit(x_train, y_train) 

# print best parameter after tuning 
#print(grid.best_params_) 
re_DecisionTreeClassifier=grid_DecisionTreeClassifier.cv_results_
#print(re)
grid_predictions = grid_DecisionTreeClassifier.predict(x_test) 
   
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, grid_predictions)

# print classification report 
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, grid_predictions)

from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,grid_predictions,average='weighted')

print("The confusion Matrix:\n",cm)
print("The report:\n",clf_report)
print("The f1_macro value for best parameter {}:".format(grid_DecisionTreeClassifier.best_params_),f1_macro)

from sklearn.metrics import roc_auc_score
roc_auc_score=roc_auc_score(y_test,grid_DecisionTreeClassifier.predict_proba(x_test)[:,1])
print ("roc_auc_score=",roc_auc_score)
```

```python
#RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [100, 200, 300],  
    'criterion': ['gini', 'entropy'], 
    'max_depth': [5, 10, None],  
    'max_features': ['auto', 'sqrt', 'log2'],
    'class_weight': [None, 'balanced']  
}
grid_RandomForestClassifier = GridSearchCV(RandomForestClassifier(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1') 

# fitting the model for grid search 
grid_RandomForestClassifier.fit(x_train, y_train) 

# print best parameter after tuning 
#print(grid.best_params_) 
re_RandomForestClassifier=grid_RandomForestClassifier.cv_results_

#print(re)
grid_predictions = grid_RandomForestClassifier.predict(x_test) 
   
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, grid_predictions)

# print classification report 
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, grid_predictions)

from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,grid_predictions,average='weighted')

print("The confusion Matrix:\n",cm)
print("The report:\n",clf_report)
print("The f1_macro value for best parameter {}:".format(grid_RandomForestClassifier.best_params_),f1_macro)

from sklearn.metrics import roc_auc_score
roc_auc_score=roc_auc_score(y_test,grid_RandomForestClassifier.predict_proba(x_test)[:,1])
print ("roc_auc_score=",roc_auc_score)
```
